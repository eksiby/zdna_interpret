```python
import numpy as np
import pandas as pd

def get_ranked_features(features_weights):
    features = features_weights
    p_deviation = pd.DataFrame()

    for column in features_weights.columns:
        if column == 'Unnamed: 0':
            continue

        mean = features_weights[column].mean()
        disp = (features_weights[column] - mean) / mean
        p_deviation[f'{column}_p_deviation'] = disp * 100

    p_deviation['mean_dev'] = p_deviation.mean(axis=1)
    p_deviation['Unnamed: 0'] = features['Unnamed: 0']
    main_p_dev = p_deviation[["Unnamed: 0",'mean_dev']]
    features_range = main_p_dev.sort_values(by='mean_dev', ascending=False)

    return features_range
int_grad = pd.read_csv("integrated_gradients_interpret_res.csv")
int_grad = int_grad.rename(columns={"impact": "int_grad"})

shap = pd.read_csv("shap_interpret_res.csv")
shap = shap.rename(columns={"impact": "shap"})

feats = shap.join(int_grad.set_index('Unnamed: 0'),on='Unnamed: 0')
get_ranked_features(feats).to_csv("ranking.csv", index=False)
```
